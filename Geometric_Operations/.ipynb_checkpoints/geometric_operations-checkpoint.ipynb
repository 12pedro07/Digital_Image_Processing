## Function definition

```python
# x = horizontal
# y = vertical

def geometry(img, translation, multiplication, circular=False, rescale=False):
    # circulate does the pixels apear in the other side of the image
    # J(x,y) = I((x+x_sum)*x_multiply, (y+y_sum)*y_multiply)

    x_sum,  y_sum  = translation
    x_mult, y_mult = multiplication

    height, width = img.shape
    if rescale:
        height = height/y_scale
        width  = width/x_scale

    J = np.zeros((height, width))

    for x in range(len(J)):
        for y in range(len(J[x])):
            reference_x = (x+x_sum)*x_mult
            reference_y = (y+y_sum)*y_mult
            
            if not circular:
                reference_x = None if reference_x < 0 else reference_x
                reference_y = None if reference_y < 0 else reference_y
            
            reference_x = None if reference_x > height-1 else reference_x
            reference_y = None if reference_y > width-1  else reference_y
            
            J[x][y] = img[reference_x][reference_y] if reference_x and reference_y else 0

    return J
```
