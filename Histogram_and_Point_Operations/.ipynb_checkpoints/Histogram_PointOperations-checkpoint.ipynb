```python
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt

# global variables
absolute_path = os.path.join(os.path.dirname(os.getcwd()), 'imgs', 'radiografia.png');
```

```python
# Functions

def histogram(img):
    h = [0]*256
    for i in range(len(img)):
        for j in range(len(img[0])):
            h[img[i][j]] += 1
    return h

def cumulative_frequency(img):
    cdf = histogram(img)
    c = 0
    total = sum(h)
    for idx, value in enumerate(cdf):
        cdf[idx] = (value + c) / total
        c += value
    return cdf

def thresholding(img, T):
    # F(D) = step(D - T)*255
    for i in range(len(img)):
        for j in range(len(img[i])):
            if img[i][j] > T:
                img[i][j] = 255
            else:
                img[i][j] = 0
                
def negative(img):
    # F(D) = 255-D
    for i in range(len(img)):
        for j in range(len(img[i])):
            img[i][j] = 255-img[i][j]
            
def contrast_stretching(img,L,H):
    #F(D) = 255/(H-L)*(D-L)
    for i in range(len(img)):
        for j in range(len(img[i])):
            img[i][j] = 255/(H-L)*(img[i][j]-L)
            if img[i][j] > 255: img[i][j] = 255
            elif img[i][j] < 0: img[i][j] = 0

def hist_equalization(img):
    # non-linear rearrangement of the histogram
    # treats the histogram as a probability mass function
    # use cdf (cumulative density function)        
    
    cdf = cumulative_frequency(img)
    for i in range(len(img)):
        for j in range(len(img[i])):
            img[i][j] = cdf[img[i][j]] * 255
    
```

```python
# loading the original image
img = cv2.imread(absolute_path,0)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Creating a histogram (and cdf) of the image

# manual
h   = histogram(img)
cdf = cumulative_frequency(img)
x   = [i for i in range(256)]

# opencv / np
h_cv2  = cv2.calcHist([img],[0],None,[256],[0,256]) 
cdf_np = np.cumsum(h_cv2)/np.sum(h_cv2)

# plots
plt.figure(0,figsize=(15,5))

# manual
plt.subplot(221)
plt.title("manual histogram / cdf")
plt.bar(x,h, width=1)
plt.subplot(223)
plt.plot(x,cdf)

# opencv / np
plt.subplot(222)
plt.title("opencv / numpy")
plt.plot(h_cv2)
plt.subplot(224)
plt.plot(cdf_np)

plt.show()
```

```python
# Testing thresholding

# manual
thresholded = img.copy()
thresholding(thresholded,45)
plt.figure(1,figsize=(15,5))
plt.subplot(121)
plt.title("manual")
plt.imshow(thresholded, cmap='gray')

# opencv
thresholded_cv2 = img.copy()
ret, thresholded_cv2 = cv2.threshold(thresholded_cv2,45,255,cv2.THRESH_BINARY)
plt.subplot(122)
plt.title("opencv")
plt.imshow(thresholded_cv2, cmap='gray')
plt.show()
```

```python
# Testing negative

plt.figure(1,figsize=(15,5))

# manual
neg_img = img.copy()
negative(neg_img)
plt.subplot(121)
plt.title("manual")
plt.imshow(neg_img, cmap='gray')

# opencv
neg_cv2 = cv2.bitwise_not(img.copy())
plt.subplot(122)
plt.title("opencv")
plt.imshow(neg_cv2, cmap='gray')
plt.show()
```

```python
# Testing contrast stretching

# from the histogram we know that the values are contained from 0 to 75
cont_img = img.copy()
contrast_stretching(cont_img,0,75)
plt.figure(1,figsize=(10,4.5))
plt.imshow(cont_img, cmap='gray')
plt.title("manual")
plt.show()

# opencv doesn't have a method to do contrast stretching
```

```python
# Testing histogram equalization

plt.figure(1,figsize=(15,10))

# manual
eq_img = img.copy()
hist_equalization(eq_img)
plt.subplot(121)
plt.title("manual")
plt.imshow(eq_img, cmap='gray')

# opencv
eq_cv2 = cv2.equalizeHist(img)
plt.subplot(122)
plt.title("opencv")
plt.imshow(eq_cv2, cmap='gray')
plt.show()


# plots
h_eq    = histogram(eq_img)
cdf_eq  = cumulative_frequency(eq_img)
h_cv2   = cv2.calcHist([eq_cv2],[0],None,[256],[0,256])
cdf_np  = np.cumsum(h_cv2)/np.sum(h_cv2)
x       = [i for i in range(256)]
plt.figure(2,figsize=(15,5))

# manual
plt.subplot(221)
plt.bar(x,h, width=1)
plt.subplot(223)
plt.plot(x,cdf)
plt.subplot(222)
plt.plot(h_cv2)
plt.subplot(224)
plt.plot(cdf_np)
```

```python
# Some notes about gamma correction

# Problem: Displays have a gamma factor, distorting the output luminance intensities from the desired one.
# In other words we have Output(I) = Input(I)^gamma.

Input = [i/1000 for i in range(1000)]
Output_lower  = [(i**0.25) for i in Input]
Output_ideal  = [(i**1.0) for i in Input]
Output_higher = [(i**3) for i in Input]

plt.plot(Input,Output_lower,  color='red',   label='gamma < 1')
plt.plot(Input,Output_ideal,  color='blue',  label='gamma = 1')
plt.plot(Input,Output_higher, color='green', label='gamma > 1')
plt.ylim(0, 1)
plt.xlabel("Input (%)")
plt.xlim(0, 1)
plt.ylabel("Output (%)")
plt.legend()
plt.show()

# Solution: Compensate the gamma factor Output(I) = (Input(I)^gamma)^(1/gamma)
```
